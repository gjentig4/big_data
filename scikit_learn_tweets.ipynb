```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
#Uncomment the line below when using the whole dataset at once
#from sklearn.model_selection import train_test_split
#Lib to divide the training data1
from sklearn.model_selection import KFold

# Load the cleaned tweet data into a DataFrame
tweets = pd.read_csv('preprocessed_tweets.csv')
```

```python

# # Split the data into train and test sets
# X_train, X_test, y_train, y_test = train_test_split(tweets['clean_tweet'], tweets['pola'], test_size=0.2, random_state=42)

# # Vectorize the tweets using TF-IDF
# vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
# X_train_vec = vectorizer.fit_transform(X_train)
# X_test_vec = vectorizer.transform(X_test)

# # Train a KNN classifier with k=5
# knn = KNeighborsClassifier(n_neighbors=5)
# knn.fit(X_train_vec, y_train)

# # Predict the sentiment of the test set
# y_pred = knn.predict(X_test_vec)

# # Calculate the accuracy of the classifier
# accuracy = accuracy_score(y_test, y_pred)
# print(f'Accuracy: {accuracy:.2f}')

#------------------------------------------------------

# Vectorize (transform to numbers) the tweets using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
X = vectorizer.fit_transform(tweets['clean_tweet'])
y = tweets['pola']

# Train a KNN classifier with k=5 using 5-fold cross-validation
knn = KNeighborsClassifier(n_neighbors=10)
kf = KFold(n_splits=5)
accuracy_scores = []
for train_idx, test_idx in kf.split(X):
    X_train, y_train = X[train_idx], y[train_idx]
    X_test, y_test = X[test_idx], y[test_idx]
    knn.fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    accuracy_scores.append(accuracy)

# Calculate the mean accuracy across all folds
mean_accuracy = sum(accuracy_scores) / len(accuracy_scores)
print(f'Mean accuracy: {mean_accuracy:.2f}')
```
